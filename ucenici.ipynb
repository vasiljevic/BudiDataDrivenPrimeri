```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
t = pd.read_csv('ucenici.csv')
```

```python
t.head()
```

```python
tsum = t[["Подручје рада", "Број ученика - први разред", "Број девојчица - први разред"]].groupby("Подручје рада").sum()
```

```python
tsum
```

```python
tproc = (tsum["Број девојчица - први разред"]/tsum["Број ученика - први разред"]*100).sort_values()
```

```python
tproc
```

```python
tproc.plot.barh(figsize=(12, 5))
```
