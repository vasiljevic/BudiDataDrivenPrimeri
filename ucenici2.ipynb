```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
t = pd.read_csv('ucenici.csv')
```

```python
t.head()
```

```python
tsum = t[["Образовни профил", "Број ученика - први разред", "Број девојчица - први разред"]].groupby("Образовни профил").sum()
```

```python
tsum
```

```python
tproc = (tsum["Број девојчица - први разред"]/tsum["Број ученика - први разред"]*100).sort_values().dropna()
```

```python
tproc
```

```python
tproc.plot.barh(figsize=(12, 55))
```
